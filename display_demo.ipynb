```python
%matplotlib inline
import os
import subprocess
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.animation
import nibabel as nib
```

```python
head_nifti = nib.load('out/anatomical.nii.gz')
head = head_nifti.get_data()
banana_nifti = nib.load('out/banana.nii.gz')
banana = banana_nifti.get_data()
```

```python
slicer = slice(30, 80), 100, slice(30, 90)
```

```python
plt.imshow(head[slicer])
```

```python
plt.imshow(np.log(banana[slicer]))
```

```python
cmap = plt.cm.magma
my_cmap = cmap(np.arange(cmap.N))
alphas = np.logspace(-1, 0, num=cmap.N)
my_cmap[:,-1] = alphas
my_cmap = mpl.colors.ListedColormap(my_cmap)
```

```python
plt.imshow(head[slicer], cmap='gray')
plt.imshow(np.log(banana[slicer]), cmap=my_cmap)
```
